## Filtro NLM para la eliminacin de ruido de las imgenes

```python
os.chdir(r"C:/Users/KlarSC/Desktop/Laboratorio1P3")
images_path = "./Dataset_copia_imagenes"
class_names = os.listdir(images_path)
class_names.sort()
print(class_names)
```

```python
for class_name in class_names:
    cur_path = images_path + "/" + class_name
    cur_label = class_name
    i = 1

    for file in glob.glob(cur_path+"/*"):
        print("Processing Image - {} in {}".format(i, cur_label))
        # read the images path
        image = cv2.imread(file)
        # denoising images
        #convertir a 8 bits
        #image=np.uint8(img)
        dst = cv2.fastNlMeansDenoisingColored(image,None,10,10,7,21)
        b,g,r = cv2.split(dst)
        rgb_dst = cv2.merge([r,g,b])
        # save images
        save_path = './Dataset_sin_ruido/'+ cur_label
        cv2.imwrite(os.path.join(save_path , cur_label + '_' + str(i) + '.jpg'),
                    cv2.cvtColor(rgb_dst,cv2.COLOR_RGB2BGR))
        
        # show loop update
        i += 1
```

## Eliminacin de fondo en imagenes

```python
os.chdir(r"C:/Users/KlarSC/Desktop/Laboratorio1P3")
images_path = "./Dataset_sin_ruido"
class_names = os.listdir(images_path)
class_names.sort()
print(class_names)
```

```python
for class_name in class_names:
    cur_path = images_path + "/" + class_name
    cur_label = class_name
    i = 1

    for file in glob.glob(cur_path + "/*"):
        print("Processing Image - {} in {}".format(i, cur_label))
        # read the images path
        image = cv2.imread(file)
        # remove bg 
        output = remove(image)
        # save images
        save_path = './Dataset_sin_fondo/'+ cur_label
        cv2.imwrite(os.path.join(save_path , cur_label + '_' + str(i) + '.png'), output)
        

        # show loop update
        i += 1
```
